```python
# dependencies
import pandas as pd
```

```python
# get all the csv filepaths and create a list of files
bikedata_2019_01 = "data/JC-201901-citibike-tripdata.csv"
bikedata_2019_02 = "data/JC-201902-citibike-tripdata.csv"
bikedata_2019_03 = "data/JC-201903-citibike-tripdata.csv"
bikedata_2019_04 = "data/JC-201904-citibike-tripdata.csv"
bikedata_2019_05 = "data/JC-201905-citibike-tripdata.csv"
bikedata_2019_06 = "data/JC-201906-citibike-tripdata.csv"
bikedata_2019_07 = "data/JC-201907-citibike-tripdata.csv"
bikedata_2019_08 = "data/JC-201908-citibike-tripdata.csv"
bikedata_2019_09 = "data/JC-201909-citibike-tripdata.csv"
bikedata_2019_10 = "data/JC-201910-citibike-tripdata.csv"
bikedata_2019_11 = "data/JC-201911-citibike-tripdata.csv"
bikedata_2019_12 = "data/JC-201912-citibike-tripdata.csv"

all_files = [bikedata_2019_01,bikedata_2019_02,bikedata_2019_03,bikedata_2019_04,bikedata_2019_05,bikedata_2019_06,
             bikedata_2019_07, bikedata_2019_08,bikedata_2019_09,bikedata_2019_10,bikedata_2019_11,bikedata_2019_12]
```

```python
# create empty list to add dataframes to
bikedata_list = []

# for loop to go through each file and read into dataframe, then append to the new list
for file in all_files:
    bikedata_list.append(pd.read_csv(file))

# concatenate the list of dataframes to a single dataframe
bikedata_2019 = pd.concat(bikedata_list, sort=False)
bikedata_2019
```

```python
bikedata_2019.dtypes
```

```python
# clean the df to have only the data with birth year over 1919 (under 100 years old)
bikedata_df = bikedata_2019.loc[bikedata_2019["birth year"] > 1919, :]

# reset index
bikedata_df = bikedata_df.reset_index(drop = True)
```

```python
bikedata_df.describe()
```

```python
# output the final dataframe to a csv file for using in tableau

bikedata_df.to_csv("data/bikedata_2019.csv", index=False)
```
